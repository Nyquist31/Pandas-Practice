```python
# Import the car sales DataFrame with missing data (".../data/car-sales-missing-sata.csv")
car_sales_missing = pd.read_csv("car-sales-extended-missing-data.csv")
# Check the car sales DataFrame
car_sales_missing
```

```python
# Remove the redst of the missing data inplace
car_sales_missing = car_sales_missing.dropna()
# Verify the missing values are removed by viewing the DataFrame
car_sales_missing
```

```python
# Create a Seats column where every row has a value of 5
seats = pd.Series([5])
car_sales["Seats"] = seats
car_sales["Seats"] = car_sales["Seats"].fillna(car_sales["Seats"].mean())
car_sales
```

```python
# Create a column called "Engine Size" with random values between 1.3 and 4.5
# Remember: If you're doing it from a Python list,the list has to be the same length as the DataFrame
engine_column = pd.Series([1.5, 3.6, 2.7, 4.1, 2.5, 1.7, 3.5, 4, 2.3, 3.2])
car_sales["Engine Size"] = engine_column
car_sales
```

```python
# Create a column which represents the price of a car per kilometer
car_sales["Price per KM"] = car_sales["Price"] / car_sales["Odometer (KM)"]
# Then view the DataFrame
car_sales
```

```python
# Remove the las column you added using .drop()
car_sales.drop(["Price per KM"],axis =1)
```

```python
# Shuffle the DataFrame using sample() with the frac parameter set to 1
# Save the shuffled DataFrame to a new variable
car_sales_shuffled = car_sales.sample(frac = 1)
car_sales_shuffled
```

```python
# Reset the indexes of the shuffled DataFrame
car_sales_shuffled.reset_index(drop = True, inplace =True)
car_sales_shuffled
```

```python
# Change the Odometer values from kilometers to miles using a Lambda function
car_sales["Odometer (KM)"] = car_sales["Odometer (KM)"].apply(lambda x: x/1.6)
# Then view the DataFrame
car_sales
```
